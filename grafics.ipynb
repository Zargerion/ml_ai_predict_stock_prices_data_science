```python
import numpy as np
x1 = np.array([0,1,2,3,4,5,6])
x2 = np.array([0,1,4,6,8,10,12])

x1, x2
```

```python
from matplotlib import pyplot as plt

plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots()

ax.plot(x1, x2, color='blue', linestyle='-', marker='o')
ax.plot(x1, np.sin(x1)+1, color=(1.0, 0.2, 0.3), linestyle='--')
ax.fill_between(x1, np.sin(x1)+1, color=(1.0, 0.2, 0.3), alpha=0.4)

ax.set_title('USDTBTC')
ax.set_xlabel('Time')
ax.set_ylabel('Price')
ax.set_yticks(np.arange(0, 14, 1))

```

```python
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X=np.arange(-5, 5, 0.25)
Y=np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

```
